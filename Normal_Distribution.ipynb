##Generating Samples from a Normal Distribution

```python
# IMPORT REQUIRED LIBRARIES
# for statistical functionality
from scipy.stats import norm 
# for array definition
import numpy as np
# for plotting
import matplotlib.pyplot as plt
# GENERATE 10000 SAMPLES from a normal distribution with mean 0 and variance 1
data = np.random.normal(loc=10.0, scale=3.0, size=10000)
# PLOT
# what are we plotting and how many bins?
plt.hist(data, bins=40)
# what is plotted on the x-axis?
plt.xlabel('X')
# what is plotted on the y-axis?
plt.ylabel('Y')
# Overall title of the plot
plt.title('Simulated Normal Distribution')
# display the resulting plot
plt.show()
```

##Investigating different values for the mean

```python
# define figure to have multiple subplots
fig, ax = plt.subplots()
# generate 100 points between -3 and 3
x = np.linspace(-3,3,100)
# define different means
mus = [-1.0, 0.0, 1.0]
# for each of the means in mus
for i in mus:
    # plot the resulting normal distribution
    ax.plot(x, norm.pdf(x,loc=i), label='mean=%.1f'%i)
    
ax.set_xlabel('x')
ax.set_ylabel('density')
ax.set_title('Normal Distribution')
ax.legend(loc='best', frameon=True)
ax.set_ylim(0,0.45)
ax.grid(True)
plt.show()
```

##Investigating different values for standard deviation

```python
# define figure to have multiple subplots
fig, ax = plt.subplots()
# generate 100 points between -3 and 3
x = np.linspace(-3,3,100)
# define different standard deviations
stds = [1.0, 2.0, 3.0]
# each of the standard deviation in stds
for i in stdvs:
    # plot the resulting normal distribution
    ax.plot(x, norm.pdf(x,scale=i), label='std=%.1f'%i)
    
ax.set_xlabel('x')
ax.set_ylabel('density')
ax.set_title('Normal Distribution')
ax.legend(loc='best', frameon=True)
ax.set_ylim(0,0.45)
ax.grid(True)
plt.show()
```

##Plotting probabilities

```python
import scipy.stats as stats
import scipy.stats
from scipy.stats import norm
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
def draw_normdist(mu, sigma, z_low, z_high):
# define constants
    #mu = 1000
    #sigma = 100
    #x1 = 900
    #x2 = 1100
    high_prob = stats.norm.cdf((z_high-mu)/sigma)
    low_prob = stats.norm.cdf((z_low-mu)/sigma)
    prob = round(high_prob - low_prob, 4)
    # calculate the z-transform
    z1 = ( z_low - mu ) / sigma
    z2 = ( z_high - mu ) / sigma
    x = np.arange(z_low, z_high, 0.001) # range of x in spec
    x_all = np.arange(mu-3*sigma, mu+3*sigma, 0.001) # entire range of x, both in and out of spec
    # mean = 0, stddev = 1, since Z-transform was calculated
    y = norm.pdf(x, mu, sigma)
    y2 = norm.pdf(x_all, mu, sigma)
    # build the plot
    fig, ax = plt.subplots(figsize=(9,6))
    plt.style.use('fivethirtyeight')
    ax.plot(x_all,y2)
    ax.fill_between(x,y,0, alpha=0.3, color='b')
    ax.fill_between(x_all,y2,0, alpha=0.1)
    xlim_low = mu - 3*sigma
    xlim_high = mu + 3*sigma
    ax.set_xlim([xlim_low, xlim_high])
    ax.set_yticklabels([])
    if z_low < mu - 3*sigma:
        title = 'z < ' + str(z2)
    elif z_high > mu + 3*sigma:
        title = str(z1) + ' < z'
    else:
        title = str(z1) + ' < z < ' + str(z2)
    title = title + '\nprobability = ' + str(prob)
    ax.set_title(title)
    #ax.set_title(str(z_low) + ' < z < ' + str(z_high) + '\n' + 'probability = ' + str(prob))
    plt.savefig('normal_curve.png', dpi=72, bbox_inches='tight')
    plt.show()

#Example of how to use the function
draw_normdist(1000, 100, 850, 1175)
```
